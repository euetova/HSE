Множества - особенная структура данных, чем-то похожая и на массив, и на словарь, но от них отличающаяся.

Как и массивы, множества содержат элементы. Но в отличие от массивов, где элементы упорядочены, в множестве они идут в "произвольном" порядке (для компьютера он, конечно, не произвольный, а обусловленный рядом технических причин, для человека не очевидных).

На словари множества похожи тем, что, как и ключи в словарях, элементы множеств должны быть уникальны, и не следуют в фиксированном порядке. 

```python
# вот так можно превратить массив в множство:

a = [1, 2, 3, 1, 2, 3]
s = set(a)
print (s)
```

## Зачем нужны множества

Во-первых, поиск в множестве происходит быстрее, чем в массиве. Если вам нужно собрать какие-то данные, а потом проверять, есть ли в этих данных что-то, то целесообразно использовать именно множества

```python
# вариант не очень:
a = [1, 2, 3, 1, 2, 3]
if 1 in a:
    print('True')
    
# как надо:
a = [1, 2, 3, 1, 2, 3]
s = set(a)
if 1 in a:
    print('True')
    
# конечно, на таком маленьком наборе разница незаметна, но когда объём данных вырастет, 
# программы, в которых используются множства, начнут работать ощутимо быстрее.
```

Во-вторых, преобразование в множества помогают быстро превратить набор данных в набор уникальных элементов, как это было в примере выше: в массиве повторялись числа 1, 2 и 3. В множестве каждое из чисел осталось в единственном экземпляре.

## Собственно операции

Множества поддерживают разные специальные операции, которых нельзя так же просто осуществить с другими структурами данных:

```python
a = set([1, 2, 3, 4, 5, 6])
b = set([4, 5, 6, 7, 8, 9])

# объединение
c = a | b
print(c)

# пересечение
c = a & b
print(c)

# разность
c = a - b
print(c)

# симметрическая разность, то есть элементы, входящие в a или в b, но не в оба множества одновременно
c = a ^ b
print(c)
```
