```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
exp = sts.expon(loc=0.5, scale=1.2) #любимое распределение
sample = exp.rvs(1000)
```

```python
x = np.linspace(0,10,100)
pdf = exp.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.hist(sample, normed=True)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
def gr(sample,smpl_size):
    n = 1000
    means = [] 
    for i in range(n):
        means.append(np.mean(sample[np.random.randint(0,len(sample),smpl_size)]))
        
    mu = np.mean(means)                #среднее
    sigma = np.std(means,ddof=1)       # сигма
    se = sigma/np.sqrt(smpl_size)     #стандартная ошибка
    norm_rv = sts.norm(mu,se)         #нормальное распределение
    plt.hist(means,normed=True)     #строим графики
    plt.plot(x, norm_rv.pdf(x),label='theoretical pdf')
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
```

```python
gr(sample, 5)
```

```python
gr(sample, 10)
```

```python
gr(sample, 50)
```
