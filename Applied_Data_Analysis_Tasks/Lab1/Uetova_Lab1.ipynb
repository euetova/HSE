```python
import nltk
from nltk.tokenize import WhitespaceTokenizer
from string import punctuation
exclude = set(punctuation + '0123456789'+'–—'+'«»')
mumi = ''.join(ch for ch in mumi if ch not in exclude)
tokens_mumi = WhitespaceTokenizer().tokenize(mumi.lower())
for i in tokens_mumi[:10]: 
    print (i) 
```

```python
print ('N of tokens_mumi: ', len(tokens_mumi))
types_mumi = nltk.FreqDist(tokens_mumi)
print ('N of types_mumi:', len(types_mumi))
print (types_mumi)
```

```python
ugkod = ''.join(ch for ch in ugkod if ch not in exclude)
tokens_ugkod = WhitespaceTokenizer().tokenize(ugkod.lower())
for i in tokens_ugkod[:10]: 
    print (i) 
```

```python
print ('N of tokens_ugkod: ', len(tokens_ugkod))
types_ugkod = nltk.FreqDist(tokens_ugkod)
print ('N of types_ugkod:', len(types_ugkod))
print (types_ugkod)
```

```python
for i in types_mumi.most_common(20):
     print (i[0], i[1])
```

```python
for i in types_ugkod.most_common(20):
     print (i[0], i[1])
```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
lemmata_mumi = nltk.FreqDist()
for t in types_mumi:
    try:
        l = morph.parse(t)[0].normal_form
        if l in lemmata_mumi:
            lemmata_mumi[l] += types_mumi[t]
        else:
            lemmata_mumi[l] = types_mumi[t]
    except IndexError:
        if t in lemmata_mumi:
            lemmata_mumi[t] += types_mumi[t]
        else:
            lemmata_mumi[t] = types_mumi[t]
print ('N of lemmata_mumi:', len(lemmata_mumi))
for i in lemmata_mumi.most_common(20):
    print (i[0], i[1])
```

```python
lemmata_ugkod = nltk.FreqDist()
for t in types_ugkod:
    try:
        l = morph.parse(t)[0].normal_form
        if l in lemmata_ugkod:
            lemmata_ugkod[l] += types_ugkod[t]
        else:
            lemmata_ugkod[l] = types_ugkod[t]
    except IndexError:
        if t in lemmata_ugkod:
            lemmata_ugkod[t] += types_ugkod[t]
        else:
            lemmata_ugkod[t] = types_ugkod[t]
print ('N of lemmata_ugkod:', len(lemmata_ugkod))
for i in lemmata_ugkod.most_common(20):
    print (i[0], i[1])
```

```python
pos_mumi = nltk.FreqDist()
for t in types_mumi:
    try:
        gram_info = morph.parse(t)[0]
        l = gram_info.tag.POS
        if l in pos_mumi:
            pos_mumi[l] += types_mumi[t]
        else:
            pos_mumi[l] = types_mumi[t]
    except IndexError:
        pass
for i in pos_mumi.most_common(20):
    print (i[0], i[1])
```

```python
pos_ugkod = nltk.FreqDist()
for t in types_ugkod:
    try:
        gram_info = morph.parse(t)[0]
        l = gram_info.tag.POS
        if l in pos_ugkod:
            pos_ugkod[l] += types_ugkod[t]
        else:
            pos_ugkod[l] = types_ugkod[t]
    except IndexError:
        pass
for i in pos_ugkod.most_common(20):
    print (i[0], i[1])
```

```python
res=list(set(pos_mumi.keys()) & set(pos_ugkod.keys()))
```

```python
for i in range(len(res)):
    res[i] = str(res[i])
```

```python
res = sorted(res)
```

```python
nom_mumi = []
for i in res:
    print(i)
    nom_mumi.append(pos_mumi[i])
print(nom_mumi)
```

```python
nom_ugkod = []
for i in res:
    print(i)
    nom_ugkod.append(pos_ugkod[i])
print(nom_ugkod)
```

```python
from scipy import stats
stats.spearmanr(nom_mumi, nom_ugkod)
```
